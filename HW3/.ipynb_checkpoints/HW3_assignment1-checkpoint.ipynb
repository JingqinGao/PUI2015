```python
from __future__ import print_function
__author__='JG'

import pylab as pl
import numpy as np
%pylab inline

import json
import os
s = json.load( open('fbb_matplotlibrc.json') )
pl.rcParams.update(s)
```

```python
#set up the population mean as 100
mymean = 100
distributions = {}
#Create 100 samples of different sizes
mysizes=2000/(np.array(range(1,100)+[10]))
#range returns a list not an array
#print (mysizes)
```

## GENERATE  100 samples of different sizes for six distributions with the same population mean. 
##1. Chi-square Distribution

```python
#numpy.random.chisquare(df, size=None)
#mean = df (number of degrees of freedom)
#set up a random seed
numpy.random.seed(seed=1111)
df = mymean
distributions={'chisq':{}}
for n in mysizes:
    distributions['chisq'][n] = np.random.chisquare(df,n)
```

```python
means={}
means['chisq'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['chisq']: # or use distributions['chisq'].iterkeys()
    means['chisq'][nn] = distributions['chisq'][nn].mean()
    #plot it
    axchisq_mu_n.plot(nn, distributions['chisq'][nn].mean(), 's')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysizes),max(mysizes)],[df,df], 'k')
    axchisq_mu_n.set_title('Sample Means of Chi-Square Distributions', fontsize=20, fontweight='bold')
```

###Question:
###Describe the behavior you see in the plots - do they look as you expected? why?

I'm expecting the samples will have different means, but the sample means should to be close to the population mean (form a normal shape with an expected value of the population mean 100)

As expected, the distribution of sample means tends to have a normal distribution with a mean value of 100.  



##2. Normal Distribution

```python
#Normal Distribution
#numpy.random.normal(loc=0.0, scale=1.0, size=None)
#mean = location
numpy.random.seed(seed=1111)
loc = mymean
distributions={'normal':{}}
for n in mysizes:
    distributions['normal'][n] = np.random.normal(loc,20,n)
```

```python
means['normal'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axnormal_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['normal']: 
    means['normal'][nn] = distributions['normal'][nn].mean()
    #plot it
    axnormal_mu_n.plot(nn, distributions['normal'][nn].mean(), 'o')
    axnormal_mu_n.set_xlabel('sample size')
    axnormal_mu_n.set_ylabel('sample mean')
    axnormal_mu_n.plot([min(mysizes),max(mysizes)],[loc,loc], 'k')
    axnormal_mu_n.set_title('Sample Means of Normal Distributions', fontsize=20, fontweight='bold')

#a=len(means['norm'])
#print (a)
```

##3. Poisson Distribution

```python
#Poisson Distribution
#numpy.random.poisson(lam=1.0, size=None)
#mean = lamda
numpy.random.seed(seed=1111)
lam = mymean
distributions={'poisson':{}}
for n in mysizes:
    distributions['poisson'][n] = np.random.poisson(lam,n)
```

```python
means['poisson'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axpoisson_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['poisson']: 
    means['poisson'][nn] = distributions['poisson'][nn].mean()
    #plot it
    axpoisson_mu_n.plot(nn, distributions['poisson'][nn].mean(), '>')
    axpoisson_mu_n.set_xlabel('sample size')
    axpoisson_mu_n.set_ylabel('sample mean')
    axpoisson_mu_n.plot([min(mysizes),max(mysizes)],[lam,lam], 'k')
    axpoisson_mu_n.set_title('Sample Means of Poisson Distributions', fontsize=20, fontweight='bold')
```

#4. Binomial Distribution

```python
#Binomial Distribution
#numpy.random.binomial(n, p, size=None)
#mean = np
numpy.random.seed(seed=1111)
my_p = 0.5
my_n = mymean/my_p
mean_b = my_n*my_p
distributions={'binomial':{}}
for n in mysizes:
    distributions['binomial'][n] = np.random.binomial(my_n,my_p,n)
```

```python
means['binomial'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axbinomial_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['binomial']: 
    means['binomial'][nn] = distributions['binomial'][nn].mean()
    #plot it
    axbinomial_mu_n.plot(nn, distributions['binomial'][nn].mean(), '*')
    axbinomial_mu_n.set_xlabel('sample size')
    axbinomial_mu_n.set_ylabel('sample mean')
    axbinomial_mu_n.plot([min(mysizes),max(mysizes)],[mean_b,mean_b], 'k')
    axbinomial_mu_n.set_title('Sample Means of Binomial Distributions', fontsize=20, fontweight='bold')
```

#5. Triangular Distribution

```python
#Triangular Distribution
#numpy.random.triangular(left, mode, right, size=None)
#mean=(left+mode+right)/3
numpy.random.seed(seed=1111)
myleft=60
mymode=60
myright=180
mean_t=(myleft+mymode+myright)/3
distributions={'triangular':{}}
for n in mysizes:
    distributions['triangular'][n] = np.random.triangular(myleft,mymode,myright,n)
```

```python
means['triangular'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axtriangular_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['triangular']: 
    means['triangular'][nn] = distributions['triangular'][nn].mean()
    #plot it
    axtriangular_mu_n.plot(nn, distributions['triangular'][nn].mean(), 'd')
    axtriangular_mu_n.set_xlabel('sample size')
    axtriangular_mu_n.set_ylabel('sample mean')
    axtriangular_mu_n.plot([min(mysizes),max(mysizes)],[mean_t,mean_t], 'k')
    axtriangular_mu_n.set_title('Sample Means of Triangular Distributions', fontsize=20, fontweight='bold')
```

#6. Exponential Distribution

```python
#Exponential Distribution
#numpy.random.exponential(scale=1.0, size=None)
#mean=scale=1/lamda
numpy.random.seed(seed=1111)
myscale=mymean
distributions={'exponential':{}}
for n in mysizes:
    distributions['exponential'][n] = np.random.exponential(myscale,n)
```

```python
means['exponential'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axexponential_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in distributions['exponential']: 
    means['exponential'][nn] = distributions['exponential'][nn].mean()
    #plot it
    axexponential_mu_n.plot(nn, distributions['exponential'][nn].mean(), '^')
    axexponential_mu_n.set_xlabel('sample size')
    axexponential_mu_n.set_ylabel('sample mean')
    axexponential_mu_n.plot([min(mysizes),max(mysizes)],[myscale,myscale], 'k')
    axexponential_mu_n.set_title('Sample Means of Exponential Distributions', fontsize=20, fontweight='bold')
```

#Plot the distributions of all sample means as a histogram
